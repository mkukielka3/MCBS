Build a classifier for fashion MNIST.

**1. Use exactly the same architectures (both densely connected layers and from convolutional layers) as the above MNIST** e.g., replace the dataset. Save the Jupyter Notebook in its original format and output a PDF file after training, testing, and validation. Make sure to write down how do they perform (training accuracny, testing accuracy).  

```python
import tensorflow.keras as keras
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import pandas as pd
from tensorflow.keras import layers, models
from tensorflow.keras.utils import to_categorical

tf.compat.v1.disable_eager_execution()
```

```python
(train_imgs_fash, train_labels_fash), (test_imgs_fash, test_labels_fash) = keras.datasets.fashion_mnist.load_data()
```

```python
train_images_conv = train_imgs_fash.reshape((60000, 28, 28, 1))
train_images_conv = train_images_conv.astype('float32') / 255
test_images_conv = test_imgs_fash.reshape((10000, 28, 28, 1))
test_images_conv = test_images_conv.astype('float32') / 255
```

```python
train_labels = to_categorical(train_labels_fash)
test_labels = to_categorical(test_labels_fash)
```

```python
model2 = models.Sequential()
model2.add(layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(64, (3, 3), activation='relu'))
model2.add(layers.MaxPooling2D((2, 2)))
model2.add(layers.Conv2D(64, (3, 3), activation='relu'))

model2.add(layers.Flatten())
model2.add(layers.Dense(64, activation='relu'))
model2.add(layers.Dense(10, activation='softmax'))

model2.compile(optimizer='rmsprop',
loss='categorical_crossentropy',
metrics=['accuracy'])
```

```python
model2.fit(train_images_conv, train_labels, epochs=5, batch_size=64)
```

```python
train_loss, train_acc = model2.evaluate(train_images_conv, train_labels)
test_loss, test_acc = model2.evaluate(test_images_conv, test_labels)
```

```python
#
data = {'Train':  [train_acc, train_loss], 'Test': [test_acc, test_loss]}
pd.DataFrame.from_dict(data,columns = ['Accuracy', 'Loss'], orient='index')
```

##**HOMEWORK 2**:

Write a function that takes as arguments the name of the layer and filter index and outputs the displayable filter response.  
Then you can choose different filters and visualize which patterns they are responsive too! Submit the code (as Notebook) and at least 3 filter responses (a PDF file).

```python
from tensorflow.keras.applications import VGG19
from tensorflow.keras import backend as K
import numpy as np

import matplotlib.pyplot as plt

#Load pretrained model
#we omit the densely connected layers of the network
model = VGG19(weights='imagenet', include_top=False)
model.summary()
```

```python
def get_filter_response(layer_name, filter_index, mod = model, ax = None):
  '''
  Given a *layer_name* and *filter_index* return a visual representation of
  the filter response. Optional arguments:
    *mod* - Keras model 
    *ax* - pyplot's Axes in which the response should be displayed 
  '''
  
  def deprocess_image(x):
      """Postprocess to turn into displayable image"""
      x -= x.mean()
      x /= (x.std() + 1e-5)
      x *= 0.1
      
      x += 0.5
      x = np.clip(x, 0, 1)
      
      x *= 255
      x = np.clip(x, 0, 255).astype('uint8')
      return x

  layer_output = mod.get_layer(layer_name).output

  loss = K.mean(layer_output[:, :, :, filter_index])
  grads = K.gradients(loss, mod.input)[0]

  grads /= (K.sqrt(K.mean(K.square(grads))) + 1e-5)
  iterate = K.function([model.input], [loss, grads])

  loss_value, grads_value = iterate([np.zeros((1, 150, 150, 3))])
  
  input_img_data = np.random.random((1, 150, 150, 3)) * 20 + 128.
  step = 1.

  for i in range(40):
      loss_value, grads_value = iterate([input_img_data])
      input_img_data += grads_value * step
  
  if not ax:
    plt.title(f"Layer: {layer_name}\nFilter index: {filter_index}")
    plt.imshow(deprocess_image(input_img_data[0]))
    plt.savefig(' {layer_name}_idx{filter_index}.pdf')
  else:
    ax.title.set_text(f"Layer: {layer_name}\nFilter index: {filter_index}")
    ax.imshow(deprocess_image(input_img_data[0]))
```

```python
get_filter_response('block1_conv2', 5)
```

```python
# Saving plots of filter responses to PDF file 
fig, axs = plt.subplots(2,3)
fig.set_size_inches(20, 15)
fig.suptitle("Filter responses", fontsize = 25)

response_lst = [('block4_conv2', 20), ('block2_conv2', 50),('block4_conv2', 5),
                ('block5_conv3', 10), ('block5_conv1', 500),('block1_conv1', 60)]

for i, values in enumerate(response_lst):
    ax_x, ax_y = i//3, i- i//3*3
    layer, filter_idx = values
    get_filter_response(layer, filter_idx, ax= axs[ax_x, ax_y])

plt.savefig("./figures/kukielka_homework2.pdf")
```
