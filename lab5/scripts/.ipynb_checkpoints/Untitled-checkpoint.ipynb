### Homework Problem 3:
Make heatmaps of Pearson correlations statistics of samples. For example, see Figure 2 and 3 freom Gilad and Mizrahi-Man (2015) F1000Research: \url{https://f1000research.com/articles/4-121}.
First, compute the correlation statistics among columns. Second, create a heatmap using heatmap.2(). Make sure to create or add labels for samples (cancer vs. normal; batch numbers; others)

```r
load(file="../data/bottomly.Rdata")
ls()

edata <- as.matrix(exprs(bottomly.eset))

edata <- edata[rowMeans(edata) > 10, ]
edata <- log2(as.matrix(edata) + 1)
```

```r
pdata = pData(bottomly.eset)
```

```r
dim(pdata)
head(pdata)
```

```r
dim(edata)
colnames(edata)
```

```r
mod = model.matrix(~as.factor(pdata$strain),data=pdata)
mod0 = model.matrix(~1, data=pdata)
sva_output = sva(edata, mod, mod0, n.sv=num.sv(edata,mod,method="leek"))
```

```r
head(sva_output$sv)
```

```r
sva_batch <- tibble(SV1=sva_output$sv[,1],SV2=sva_output$sv[,2],batch=as.factor(pdata$experiment.number), strain=as.factor(pdata$strain))

ggplot(sva_batch) + geom_point(aes(x=SV1,y=SV2, col=strain))
```

```r
sva_batch <- gather(sva_batch,"sv","value",-batch)
```

```r
modsva = lm(t(edata) ~ as.factor(pdata$strain) + sva_output$sv)
modsva_tidy <- tidy(modsva)
```

```r
pvalues <- tibble(
  #LinearModel = mod_tidy %>% filter(term == "as.factor(pheno$cancer)Cancer") %>% select("p.value") %>% unlist,
  SVA = modsva_tidy %>% filter(term == "as.factor(pdata$strain)Strain") %>% select("p.value") %>% unlist)
```

```r
ggplot(sva_batch) + geom_violin(aes(x=batch,y=value)) + geom_jitter(aes(x=batch,y=value,col=batch)) + facet_wrap(~ sv, ncol = 1)
```

```r
library(bladderbatch)
data(bladderdata)

# sample info
pheno = pData(bladderEset)
# expression data
edata = exprs(bladderEset)
```

```r
row.variances <- apply(edata, 1, function(x) var(x))

edata <- edata[row.variances < 6,]
edata.log <- log2(edata)
```

```r
library(gplots)
library(RColorBrewer)
my_palette <- colorRampPalette(c("blue", "white", "darkred"))(n = 299)
```

```r
png("bladder.png",height=700,width=700)
heatmap.2(edata,
          main = "Bladder Cancer Data Clustered", # heat map title
          notecol="black",      # change font color of cell labels to black
          density.info="none",  # turns off density plot inside color legend
          trace="none",         # turns off trace lines inside the heat map
          margins =c(12,9),     # widens margins around plot
          col=my_palette,       # use on color palette defined earlier 
          dendrogram="none",     # only draw a row dendrogram
          scale = "row")
dev.off()
```
